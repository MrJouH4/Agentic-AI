```python
import chromadb

client = chromadb.Client()
```

```python
collection = client.create_collection("news_articles")
```

```python
collection.add(
    ids=["id1", "id2", "id3", "id4"],
    documents=[
        "Apple is leading in a smart phone game with iPhone sales up by 35%",
        "Tesla booked a minor profit of 1 billion $ in Q2",
        "Apples are high in fiber, vitamin C, and various antioxidants",
        "SpaceX got NASA contract worth 10 billion $",
    ]
)
```

```python
data = collection.get(
    include=["documents", "embeddings", "metadatas"]
)

for i, emb in enumerate(data["embeddings"]):
    print(f"\n Document {i+1}")
    print("Text: ", data["documents"][i])
    print("Embeddings length: ", len(emb))
    print("First 10 values: ", emb[:10])
```

```python
result = collection.query(query_texts=["this query related to elon musk"], n_results=2)

print(result)
```
